```python
import numpy as np
import pandas as pd
import gc

from xgboost import XGBRegressor
from scipy.stats import spearmanr
```

```python
def safe_spearman(y_true, y_pred):
    mask = ~np.isnan(y_true)

    if mask.sum() < 3:
        return np.nan

    corr = spearmanr(y_true[mask], y_pred[mask]).correlation

    if pd.isna(corr):
        return np.nan

    return corr
```

```python
targets_to_train = target_cols[:20]

valid_scores = {}
valid_predictions = {}
test_predictions = {}

for i, target in enumerate(targets_to_train):
    print(f"\nTraining {i + 1}/{len(targets_to_train)}: {target}")

    y_train = train_full.loc[train_mask, target].values
    y_valid = train_full.loc[valid_mask, target].values

    train_target_mask = ~np.isnan(y_train)
    valid_target_mask = ~np.isnan(y_valid)

    if train_target_mask.sum() < 100:
        print(f"Skipping {target}: not enough non-missing training labels.")
        valid_scores[target] = np.nan
        test_predictions[target] = np.zeros(len(X_test))
        continue

    model = XGBRegressor(
        n_estimators=300,
        learning_rate=0.03,
        max_depth=4,
        subsample=0.8,
        colsample_bytree=0.8,
        objective="reg:squarederror",
        tree_method="hist",
        device="cuda",
        random_state=42,
        n_jobs=-1,
        eval_metric="rmse"
    )

    model.fit(
        X_train.loc[train_target_mask],
        y_train[train_target_mask],
        eval_set=[
            (
                X_valid.loc[valid_target_mask],
                y_valid[valid_target_mask]
            )
        ],
        verbose=False
    )

    valid_pred = model.predict(X_valid)
    test_pred = model.predict(X_test)

    score = safe_spearman(y_valid, valid_pred)

    valid_scores[target] = score
    valid_predictions[target] = valid_pred
    test_predictions[target] = test_pred

    print(f"{target} validation Spearman: {score:.5f}")

    del model
    gc.collect()
```

```python
# ============================================================
# 13. Tune XGBoost Configurations on First 20 Targets
# Using Raw Baseline Features Only
# ============================================================

import numpy as np
import pandas as pd
import gc
import time

from xgboost import XGBRegressor
from scipy.stats import spearmanr


def safe_spearman(y_true, y_pred):
    mask = ~np.isnan(y_true)

    if mask.sum() < 3:
        return np.nan

    corr = spearmanr(y_true[mask], y_pred[mask]).correlation

    if pd.isna(corr):
        return np.nan

    return corr


# ------------------------------------------------------------
# Different model setups to test
# ------------------------------------------------------------

xgb_configs = {
    "current_baseline": {
        "n_estimators": 300,
        "learning_rate": 0.03,
        "max_depth": 4,
        "subsample": 0.8,
        "colsample_bytree": 0.8,
        "reg_alpha": 0.0,
        "reg_lambda": 1.0,
        "min_child_weight": 1,
    },

    "more_regularized": {
        "n_estimators": 500,
        "learning_rate": 0.02,
        "max_depth": 3,
        "subsample": 0.8,
        "colsample_bytree": 0.7,
        "reg_alpha": 0.1,
        "reg_lambda": 5.0,
        "min_child_weight": 5,
    },

    "shallower_simple": {
        "n_estimators": 250,
        "learning_rate": 0.04,
        "max_depth": 2,
        "subsample": 0.9,
        "colsample_bytree": 0.9,
        "reg_alpha": 0.0,
        "reg_lambda": 2.0,
        "min_child_weight": 3,
    },

    "deeper_but_regularized": {
        "n_estimators": 400,
        "learning_rate": 0.025,
        "max_depth": 5,
        "subsample": 0.75,
        "colsample_bytree": 0.75,
        "reg_alpha": 0.2,
        "reg_lambda": 8.0,
        "min_child_weight": 8,
    },
}


targets_to_train = target_cols[:20]

config_results = {}
config_target_scores = {}

start_time = time.time()

for config_name, params in xgb_configs.items():
    print("\n" + "=" * 60)
    print(f"Testing config: {config_name}")
    print("=" * 60)

    valid_scores_config = {}

    for i, target in enumerate(targets_to_train):
        print(f"Training {i + 1}/{len(targets_to_train)}: {target}")

        y_train = train_full.loc[train_mask, target].values
        y_valid = train_full.loc[valid_mask, target].values

        train_target_mask = ~np.isnan(y_train)
        valid_target_mask = ~np.isnan(y_valid)

        if train_target_mask.sum() < 100:
            valid_scores_config[target] = np.nan
            continue

        model = XGBRegressor(
            **params,
            objective="reg:squarederror",
            tree_method="hist",
            device="cuda",   # remove this line if GPU causes problems
            random_state=42,
            n_jobs=-1,
            eval_metric="rmse"
        )

        model.fit(
            X_train.loc[train_target_mask],
            y_train[train_target_mask],
            eval_set=[
                (
                    X_valid.loc[valid_target_mask],
                    y_valid[valid_target_mask]
                )
            ],
            verbose=False
        )

        valid_pred = model.predict(X_valid)
        score = safe_spearman(y_valid, valid_pred)

        valid_scores_config[target] = score

        print(f"{target}: {score:.5f}")

        del model
        gc.collect()

    score_series_config = pd.Series(valid_scores_config)

    config_target_scores[config_name] = score_series_config
    config_results[config_name] = score_series_config.mean()

    print(f"\nAverage Spearman for {config_name}: {score_series_config.mean():.5f}")


# ------------------------------------------------------------
# Compare configs
# ------------------------------------------------------------

config_results_series = pd.Series(config_results).sort_values(ascending=False)

print("\n" + "=" * 60)
print("Config comparison")
print("=" * 60)

display(config_results_series)

best_config_name = config_results_series.index[0]
best_config_params = xgb_configs[best_config_name]

print("\nBest config:")
print(best_config_name)

print("\nBest config parameters:")
print(best_config_params)

print("\nOriginal raw baseline average was:")
print(score_series.mean())
```

```python
# ============================================================
# Score existing submission.parquet against uploaded ground truth
# ============================================================

import os
import numpy as np
import pandas as pd

SOLUTION_NULL_FILLER = -999999

# ------------------------------------------------------------
# 1. Paths
# ------------------------------------------------------------

submission_path = "/kaggle/input/datasets/stephenweiler208/xg-submission-1"
ground_truth_path = "/kaggle/input/datasets/stephenweiler208/ground-truth"
competition_data_path = "/kaggle/input/competitions/mitsui-commodity-prediction-challenge"

# If the ground truth path is a folder, find the CSV inside it
if os.path.isdir(ground_truth_path):
    files = os.listdir(ground_truth_path)
    print("Files in ground truth folder:", files)

    csv_files = [f for f in files if f.endswith(".csv")]

    if len(csv_files) == 0:
        raise FileNotFoundError("No CSV file found inside the ground truth folder.")

    ground_truth_path = os.path.join(ground_truth_path, csv_files[0])

print("Using ground truth file:", ground_truth_path)


# ------------------------------------------------------------
# 2. Load submission and ground truth
# ------------------------------------------------------------

submission = pd.read_parquet(submission_path)
solution = pd.read_csv(ground_truth_path)

print("Original submission shape:", submission.shape)
print("Original solution shape:", solution.shape)

display(submission.head())
display(solution.head())


# ------------------------------------------------------------
# 3. Add date_id to submission if needed
# ------------------------------------------------------------

if "date_id" not in submission.columns:
    test = pd.read_csv(f"{competition_data_path}/test.csv")
    test_date_ids = test["date_id"].drop_duplicates().reset_index(drop=True)

    submission = submission.copy()
    submission.insert(0, "date_id", test_date_ids.iloc[:len(submission)].values)

print("Submission columns after date_id check:", submission.columns[:10].tolist())
print("Solution columns:", solution.columns[:10].tolist())


# ------------------------------------------------------------
# 4. Metric functions
# ------------------------------------------------------------

def rank_correlation_sharpe_ratio(merged_df: pd.DataFrame) -> float:
    """
    Calculates daily rank correlation between predictions and target values,
    then returns mean daily rank correlation divided by its standard deviation.
    """

    prediction_cols = [col for col in merged_df.columns if col.startswith("prediction_")]
    target_cols = [col for col in merged_df.columns if col.startswith("target_")]

    def _compute_rank_correlation(row):
        non_null_targets = [col for col in target_cols if not pd.isnull(row[col])]

        matching_predictions = [
            col for col in prediction_cols
            if col.replace("prediction", "target") in non_null_targets
        ]

        if not non_null_targets:
            raise ValueError("No non-null target values found.")

        if row[non_null_targets].std(ddof=0) == 0 or row[matching_predictions].std(ddof=0) == 0:
            raise ZeroDivisionError("Denominator is zero, unable to compute rank correlation.")

        return np.corrcoef(
            row[matching_predictions].rank(method="average"),
            row[non_null_targets].rank(method="average")
        )[0, 1]

    daily_rank_corrs = merged_df.apply(_compute_rank_correlation, axis=1)

    std_dev = daily_rank_corrs.std(ddof=0)

    if std_dev == 0:
        raise ZeroDivisionError("Denominator is zero, unable to compute Sharpe ratio.")

    sharpe_ratio = daily_rank_corrs.mean() / std_dev

    return float(sharpe_ratio)


def score(solution: pd.DataFrame, submission: pd.DataFrame, row_id_column_name: str) -> float:
    """
    Competition-style score:
    daily Spearman/rank correlation Sharpe ratio.
    """

    solution = solution.copy()
    submission = submission.copy()

    del solution[row_id_column_name]
    del submission[row_id_column_name]

    assert all(solution.columns == submission.columns), "Solution and submission columns do not match."

    submission = submission.rename(
        columns={col: col.replace("target_", "prediction_") for col in submission.columns}
    )

    solution = solution.replace(SOLUTION_NULL_FILLER, None)

    merged_df = pd.concat([solution, submission], axis="columns")

    return rank_correlation_sharpe_ratio(merged_df)


# ------------------------------------------------------------
# 5. Clean and align columns
# ------------------------------------------------------------

# Keep only target columns from the ground truth
target_cols = [c for c in solution.columns if c.startswith("target_")]

if len(target_cols) == 0:
    raise ValueError("No target_ columns found in the ground truth file.")

# Make sure submission has all required target columns
missing_targets = [c for c in target_cols if c not in submission.columns]

if missing_targets:
    raise ValueError(f"Submission is missing these target columns: {missing_targets[:10]}")

# Align rows by date_id
solution = solution[["date_id"] + target_cols].copy()
submission = submission[["date_id"] + target_cols].copy()

solution = solution.sort_values("date_id").reset_index(drop=True)
submission = submission.sort_values("date_id").reset_index(drop=True)

print("Final aligned solution shape:", solution.shape)
print("Final aligned submission shape:", submission.shape)
print("Columns match:", list(solution.columns) == list(submission.columns))

display(solution.head())
display(submission.head())


# ------------------------------------------------------------
# 6. Calculate local score
# ------------------------------------------------------------

local_score = score(
    solution=solution,
    submission=submission,
    row_id_column_name="date_id"
)

print("Local competition-style Sharpe score:")
print(local_score)
```

```python
# ============================================================
# Corrected local scoring using label_date_id alignment
# ============================================================

import os
import numpy as np
import pandas as pd

SOLUTION_NULL_FILLER = -999999

submission_path = "/kaggle/input/datasets/stephenweiler208/xg-submission-1"
ground_truth_path = "/kaggle/input/datasets/stephenweiler208/ground-truth"
competition_data_path = "/kaggle/input/competitions/mitsui-commodity-prediction-challenge"

# ------------------------------------------------------------
# Load submission
# ------------------------------------------------------------

submission = pd.read_parquet(submission_path)

test = pd.read_csv(f"{competition_data_path}/test.csv")
test_date_ids = test["date_id"].drop_duplicates().reset_index(drop=True)

if "date_id" not in submission.columns:
    submission = submission.copy()
    submission.insert(0, "date_id", test_date_ids.iloc[:len(submission)].values)

# ------------------------------------------------------------
# Load target metadata
# ------------------------------------------------------------

target_pairs = pd.read_csv(f"{competition_data_path}/target_pairs.csv")
target_cols = target_pairs["target"].tolist()

# ------------------------------------------------------------
# Rebuild solution using label_date_id, not date_id
# ------------------------------------------------------------

solution_parts = []

for lag in [1, 2, 3, 4]:
    lag_df = pd.read_csv(
        f"{competition_data_path}/lagged_test_labels/test_labels_lag_{lag}.csv"
    )

    lag_targets = target_pairs.loc[target_pairs["lag"] == lag, "target"].tolist()
    available_targets = [c for c in lag_targets if c in lag_df.columns]

    temp = lag_df[["label_date_id"] + available_targets].copy()
    temp = temp.rename(columns={"label_date_id": "date_id"})

    solution_parts.append(temp)

# Merge the different lag-specific truth pieces by the actual label date
solution = solution_parts[0]

for part in solution_parts[1:]:
    solution = solution.merge(part, on="date_id", how="outer")

# Only score dates that appear in your submission
solution = solution[solution["date_id"].isin(submission["date_id"])].copy()

# ------------------------------------------------------------
# Align rows and columns
# ------------------------------------------------------------

submission = submission[["date_id"] + target_cols].copy()
solution = solution[["date_id"] + target_cols].copy()

merged_dates = sorted(set(submission["date_id"]) & set(solution["date_id"]))

submission = submission[submission["date_id"].isin(merged_dates)].sort_values("date_id").reset_index(drop=True)
solution = solution[solution["date_id"].isin(merged_dates)].sort_values("date_id").reset_index(drop=True)

print("Scored dates:", len(merged_dates))
print("Submission shape:", submission.shape)
print("Solution shape:", solution.shape)

# ------------------------------------------------------------
# Metric
# ------------------------------------------------------------

def rank_correlation_sharpe_ratio(merged_df: pd.DataFrame) -> float:
    prediction_cols = [col for col in merged_df.columns if col.startswith("prediction_")]
    target_cols = [col for col in merged_df.columns if col.startswith("target_")]

    def _compute_rank_correlation(row):
        non_null_targets = [col for col in target_cols if not pd.isnull(row[col])]

        matching_predictions = [
            col for col in prediction_cols
            if col.replace("prediction", "target") in non_null_targets
        ]

        if not non_null_targets:
            return np.nan

        if row[non_null_targets].std(ddof=0) == 0 or row[matching_predictions].std(ddof=0) == 0:
            return np.nan

        return np.corrcoef(
            row[matching_predictions].rank(method="average"),
            row[non_null_targets].rank(method="average")
        )[0, 1]

    daily_rank_corrs = merged_df.apply(_compute_rank_correlation, axis=1).dropna()

    std_dev = daily_rank_corrs.std(ddof=0)

    if std_dev == 0:
        raise ZeroDivisionError("Denominator is zero, unable to compute Sharpe ratio.")

    return float(daily_rank_corrs.mean() / std_dev)


def score(solution: pd.DataFrame, submission: pd.DataFrame, row_id_column_name: str) -> float:
    solution = solution.copy()
    submission = submission.copy()

    del solution[row_id_column_name]
    del submission[row_id_column_name]

    assert all(solution.columns == submission.columns)

    submission = submission.rename(
        columns={col: col.replace("target_", "prediction_") for col in submission.columns}
    )

    solution = solution.replace(SOLUTION_NULL_FILLER, None)

    return rank_correlation_sharpe_ratio(
        pd.concat([solution, submission], axis="columns")
    )

local_score = score(solution, submission, row_id_column_name="date_id")

print("Corrected local competition-style Sharpe score:")
print(local_score)
```

```python
# ============================================================
# Best XGBoost Attempt: Non-Leaky + Corrected Lag Label Features
# + Rank Blend with Lag Baseline
# ============================================================

import os
import gc
import time
import warnings
import numpy as np
import pandas as pd

from xgboost import XGBRegressor

warnings.filterwarnings("ignore", category=pd.errors.PerformanceWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

SOLUTION_NULL_FILLER = -999999

competition_data_path = "/kaggle/input/competitions/mitsui-commodity-prediction-challenge"
ground_truth_path = "/kaggle/input/datasets/stephenweiler208/ground-truth"

# ============================================================
# 1. Load Ground Truth
# ============================================================

if os.path.isdir(ground_truth_path):
    csv_files = [f for f in os.listdir(ground_truth_path) if f.endswith(".csv")]
    if len(csv_files) == 0:
        raise FileNotFoundError("No CSV file found in the ground truth folder.")
    ground_truth_path = os.path.join(ground_truth_path, csv_files[0])

solution = pd.read_csv(ground_truth_path)

print("Ground truth shape:", solution.shape)
print("Ground truth date range:", solution["date_id"].min(), "to", solution["date_id"].max())

local_test_start = solution["date_id"].min()
local_test_end = solution["date_id"].max()

# ============================================================
# 2. Metadata / Target Setup
# ============================================================

target_pairs = pd.read_csv(f"{competition_data_path}/target_pairs.csv")
target_cols = target_pairs["target"].tolist()
target_lag_map = target_pairs.set_index("target")["lag"].to_dict()

print("Number of targets:", len(target_cols))
display(target_pairs.head())

# ============================================================
# 3. Make Sure Clean Feature Columns Exist
# ============================================================

id_cols = ["date_id"]
feature_cols = [c for c in train.columns if c not in id_cols]

feature_missing_rate = train[feature_cols].isna().mean()
clean_feature_cols = feature_missing_rate[feature_missing_rate <= 0.50].index.tolist()

print("Original feature count:", len(feature_cols))
print("Clean feature count:", len(clean_feature_cols))

# ============================================================
# 4. Build Corrected Lag-Label Features for TRAIN
# No fragmentation: dictionary -> DataFrame
# ============================================================

lag_feature_dict = {}

for target in target_cols:
    lag = int(target_lag_map[target])

    # Corrected alignment:
    # test lag files show lag 1 labels for label_date_id 1827 appear at date_id 1829,
    # so training uses shift(lag + 1), not shift(lag).
    lag_feature_dict[f"lag_label_{target}"] = train_labels[target].shift(lag + 1)

train_label_lag_features = pd.DataFrame(lag_feature_dict)
train_label_lag_features.insert(0, "date_id", train_labels["date_id"].values)
train_label_lag_features = train_label_lag_features.copy()

print("Train lag-label feature shape:", train_label_lag_features.shape)
display(train_label_lag_features.head())

# ============================================================
# 5. Build Lag-Label Features for TEST
# No fragmentation: merge lag groups
# ============================================================

test_label_lag_features = pd.DataFrame({"date_id": test["date_id"].values})

for lag in [1, 2, 3, 4]:
    lag_df = pd.read_csv(
        f"{competition_data_path}/lagged_test_labels/test_labels_lag_{lag}.csv"
    )

    lag_targets = target_pairs.loc[target_pairs["lag"] == lag, "target"].tolist()
    available_targets = [c for c in lag_targets if c in lag_df.columns]

    temp = lag_df[["date_id"] + available_targets].copy()
    temp = temp.rename(columns={c: f"lag_label_{c}" for c in available_targets})

    test_label_lag_features = test_label_lag_features.merge(
        temp,
        on="date_id",
        how="left"
    )

test_label_lag_features = test_label_lag_features.copy()

print("Test lag-label feature shape:", test_label_lag_features.shape)
display(test_label_lag_features.head())

# ============================================================
# 6. Build Final Feature Matrices
# ============================================================

train_features_improved = train[["date_id"] + clean_feature_cols].merge(
    train_label_lag_features,
    on="date_id",
    how="left"
)

test_features_improved = test[["date_id"] + clean_feature_cols].merge(
    test_label_lag_features,
    on="date_id",
    how="left"
)

train_features_improved = train_features_improved.replace([np.inf, -np.inf], np.nan).copy()
test_features_improved = test_features_improved.replace([np.inf, -np.inf], np.nan).copy()

improved_feature_cols = [c for c in train_features_improved.columns if c != "date_id"]

print("Improved feature count:", len(improved_feature_cols))

# ============================================================
# 7. Non-Leaky Train / Local Test Split
# ============================================================

train_nonleaky_mask = train_features_improved["date_id"] < local_test_start

X_train_improved = train_features_improved.loc[train_nonleaky_mask, improved_feature_cols].copy()
X_test_improved = test_features_improved[improved_feature_cols].copy()

print("Non-leaky training rows:", X_train_improved.shape)
print("Local test rows:", X_test_improved.shape)
print(
    "Training date range:",
    train_features_improved.loc[train_nonleaky_mask, "date_id"].min(),
    "to",
    train_features_improved.loc[train_nonleaky_mask, "date_id"].max()
)

# ============================================================
# 8. Metric Functions
# ============================================================

def rank_correlation_sharpe_ratio(merged_df: pd.DataFrame) -> float:
    prediction_cols = [col for col in merged_df.columns if col.startswith("prediction_")]
    target_cols_metric = [col for col in merged_df.columns if col.startswith("target_")]

    daily_rank_corrs = []

    for _, row in merged_df.iterrows():
        non_null_targets = [col for col in target_cols_metric if not pd.isnull(row[col])]

        matching_predictions = [
            col for col in prediction_cols
            if col.replace("prediction", "target") in non_null_targets
        ]

        if len(non_null_targets) < 3:
            daily_rank_corrs.append(np.nan)
            continue

        if row[non_null_targets].std(ddof=0) == 0 or row[matching_predictions].std(ddof=0) == 0:
            daily_rank_corrs.append(np.nan)
            continue

        corr = np.corrcoef(
            row[matching_predictions].rank(method="average"),
            row[non_null_targets].rank(method="average")
        )[0, 1]

        daily_rank_corrs.append(corr)

    daily_rank_corrs = pd.Series(daily_rank_corrs).dropna()

    std_dev = daily_rank_corrs.std(ddof=0)

    if std_dev == 0:
        raise ZeroDivisionError("Denominator is zero, unable to compute Sharpe ratio.")

    print("\nDaily rank correlation summary:")
    display(daily_rank_corrs.describe())

    print("\nMean daily rank correlation:", daily_rank_corrs.mean())
    print("Std daily rank correlation:", std_dev)

    return float(daily_rank_corrs.mean() / std_dev)


def score(solution: pd.DataFrame, submission: pd.DataFrame, row_id_column_name: str) -> float:
    solution = solution.copy()
    submission = submission.copy()

    del solution[row_id_column_name]
    del submission[row_id_column_name]

    assert all(solution.columns == submission.columns), "Solution and submission columns do not match."

    submission = submission.rename(
        columns={col: col.replace("target_", "prediction_") for col in submission.columns}
    )

    solution = solution.replace(SOLUTION_NULL_FILLER, None)

    merged = pd.concat([solution, submission], axis="columns")

    return rank_correlation_sharpe_ratio(merged)


def evaluate_submission(submission_df, solution_df, target_cols, name="model"):
    solution_eval = solution_df[["date_id"] + target_cols].copy()
    submission_eval = submission_df[["date_id"] + target_cols].copy()

    solution_eval = solution_eval.sort_values("date_id").reset_index(drop=True)
    submission_eval = submission_eval.sort_values("date_id").reset_index(drop=True)

    print("\n" + "=" * 60)
    print(f"Evaluating: {name}")
    print("=" * 60)
    print("Date IDs match:", list(solution_eval["date_id"]) == list(submission_eval["date_id"]))
    print("Columns match:", list(solution_eval.columns) == list(submission_eval.columns))

    model_score = score(
        solution=solution_eval,
        submission=submission_eval,
        row_id_column_name="date_id"
    )

    print(f"\n{name} score:", model_score)

    return model_score


# ============================================================
# 9. Train XGBoost Models
# Stronger but still regularized
# ============================================================

best_config_params = {
    "n_estimators": 350,
    "learning_rate": 0.03,
    "max_depth": 2,
    "subsample": 0.9,
    "colsample_bytree": 0.8,
    "reg_alpha": 0.05,
    "reg_lambda": 4.0,
    "min_child_weight": 4,
}

submission_xgb = pd.DataFrame(index=test.index)

start_time = time.time()

for i, target in enumerate(target_cols):
    y_train = train_labels.loc[train_nonleaky_mask, target].values
    target_mask = ~np.isnan(y_train)

    if target_mask.sum() < 100:
        submission_xgb[target] = 0.0
        continue

    model = XGBRegressor(
        **best_config_params,
        objective="reg:squarederror",
        tree_method="hist",
        random_state=42,
        n_jobs=-1,
        eval_metric="rmse"
    )

    model.fit(
        X_train_improved.loc[target_mask],
        y_train[target_mask],
        verbose=False
    )

    submission_xgb[target] = model.predict(X_test_improved)

    if (i + 1) % 25 == 0:
        elapsed = (time.time() - start_time) / 60
        print(f"Finished {i + 1}/{len(target_cols)} targets in {elapsed:.2f} minutes")

    del model
    gc.collect()

submission_xgb.insert(0, "date_id", test["date_id"].values)
submission_xgb = submission_xgb[["date_id"] + target_cols].copy()

print("\nXGBoost prediction shape:", submission_xgb.shape)
display(submission_xgb.head())

# Save XGBoost predictions
submission_xgb.to_parquet("/kaggle/working/xgb_corrected_laglabel.parquet", index=False)

# ============================================================
# 10. Lag-Label Baseline
# ============================================================

submission_lag = pd.DataFrame(index=test.index)
submission_lag.insert(0, "date_id", test["date_id"].values)

for target in target_cols:
    col = f"lag_label_{target}"
    if col in test_label_lag_features.columns:
        submission_lag[target] = test_label_lag_features[col].values
    else:
        submission_lag[target] = 0.0

submission_lag = submission_lag[["date_id"] + target_cols].copy()
submission_lag[target_cols] = submission_lag[target_cols].fillna(0.0)

print("\nLag baseline shape:", submission_lag.shape)
display(submission_lag.head())

submission_lag.to_parquet("/kaggle/working/lag_baseline.parquet", index=False)

# ============================================================
# 11. Evaluate XGBoost Alone and Lag Baseline Alone
# ============================================================

xgb_score = evaluate_submission(
    submission_xgb,
    solution,
    target_cols,
    name="XGBoost corrected lag-label"
)

lag_score = evaluate_submission(
    submission_lag,
    solution,
    target_cols,
    name="Lag-label baseline"
)

# ============================================================
# 12. Rank-Blending Search
# ============================================================

solution_eval = solution[["date_id"] + target_cols].copy()
solution_eval = solution_eval.sort_values("date_id").reset_index(drop=True)

xgb_eval = submission_xgb.sort_values("date_id").reset_index(drop=True)
lag_eval = submission_lag.sort_values("date_id").reset_index(drop=True)

xgb_rank = xgb_eval[target_cols].rank(axis=1, method="average", ascending=True)
lag_rank = lag_eval[target_cols].rank(axis=1, method="average", ascending=True)

blend_scores = {}
blend_submissions = {}

for w_xgb in [0.00, 0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80, 0.90, 1.00]:
    w_lag = 1.0 - w_xgb

    blended_values = (w_xgb * xgb_rank) + (w_lag * lag_rank)

    blended_submission = blended_values.copy()
    blended_submission.insert(0, "date_id", xgb_eval["date_id"].values)
    blended_submission = blended_submission[["date_id"] + target_cols].copy()

    blend_name = f"xgb_{w_xgb:.2f}_lag_{w_lag:.2f}"

    blend_score = evaluate_submission(
        blended_submission,
        solution,
        target_cols,
        name=blend_name
    )

    blend_scores[blend_name] = blend_score
    blend_submissions[blend_name] = blended_submission

blend_scores_series = pd.Series(blend_scores).sort_values(ascending=False)

print("\n" + "=" * 60)
print("Blend score results")
print("=" * 60)
display(blend_scores_series)

best_blend_name = blend_scores_series.index[0]
best_blend_score = blend_scores_series.iloc[0]

print("\nBest blend:", best_blend_name)
print("Best blend score:", best_blend_score)

best_submission = blend_submissions[best_blend_name]

# Save best local-eval blend
best_submission.to_parquet("/kaggle/working/submission_best_local_blend.parquet", index=False)

# Also save as submission.parquet for convenience
best_submission.to_parquet("/kaggle/working/submission.parquet", index=False)

print("\nSaved:")
print("/kaggle/working/xgb_corrected_laglabel.parquet")
print("/kaggle/working/lag_baseline.parquet")
print("/kaggle/working/submission_best_local_blend.parquet")
print("/kaggle/working/submission.parquet")

print("\nFiles in /kaggle/working:")
print(os.listdir("/kaggle/working"))
```

```python
# ============================================================
# Target-Specific Safe XGBoost Model
# Inspired by winner-style target-pair modeling
# NO negative lags, NO leakage
# ============================================================

import os
import gc
import time
import warnings
import numpy as np
import pandas as pd

from xgboost import XGBRegressor

warnings.filterwarnings("ignore", category=pd.errors.PerformanceWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

SOLUTION_NULL_FILLER = -999999

competition_data_path = "/kaggle/input/competitions/mitsui-commodity-prediction-challenge"
ground_truth_path = "/kaggle/input/datasets/stephenweiler208/ground-truth"


# ============================================================
# 1. Load / confirm ground truth
# ============================================================

if os.path.isdir(ground_truth_path):
    csv_files = [f for f in os.listdir(ground_truth_path) if f.endswith(".csv")]
    if len(csv_files) == 0:
        raise FileNotFoundError("No CSV found in ground truth folder.")
    ground_truth_path = os.path.join(ground_truth_path, csv_files[0])

solution = pd.read_csv(ground_truth_path)

local_test_start = solution["date_id"].min()
local_test_end = solution["date_id"].max()

print("Ground truth shape:", solution.shape)
print("Ground truth date range:", local_test_start, "to", local_test_end)


# ============================================================
# 2. Target metadata
# ============================================================

target_pairs = pd.read_csv(f"{competition_data_path}/target_pairs.csv")
target_cols = target_pairs["target"].tolist()
target_lag_map = target_pairs.set_index("target")["lag"].to_dict()
target_pair_map = target_pairs.set_index("target")["pair"].to_dict()

print("Number of targets:", len(target_cols))
display(target_pairs.head())


# ============================================================
# 3. Metric functions
# ============================================================

def rank_correlation_sharpe_ratio(merged_df: pd.DataFrame) -> float:
    prediction_cols = [col for col in merged_df.columns if col.startswith("prediction_")]
    target_cols_metric = [col for col in merged_df.columns if col.startswith("target_")]

    daily_rank_corrs = []

    for _, row in merged_df.iterrows():
        non_null_targets = [col for col in target_cols_metric if not pd.isnull(row[col])]

        matching_predictions = [
            col for col in prediction_cols
            if col.replace("prediction", "target") in non_null_targets
        ]

        if len(non_null_targets) < 3:
            daily_rank_corrs.append(np.nan)
            continue

        if row[non_null_targets].std(ddof=0) == 0 or row[matching_predictions].std(ddof=0) == 0:
            daily_rank_corrs.append(np.nan)
            continue

        corr = np.corrcoef(
            row[matching_predictions].rank(method="average"),
            row[non_null_targets].rank(method="average")
        )[0, 1]

        daily_rank_corrs.append(corr)

    daily_rank_corrs = pd.Series(daily_rank_corrs).dropna()
    std_dev = daily_rank_corrs.std(ddof=0)

    if std_dev == 0:
        raise ZeroDivisionError("Denominator is zero, unable to compute Sharpe ratio.")

    print("\nDaily rank correlation summary:")
    display(daily_rank_corrs.describe())
    print("Mean daily rank correlation:", daily_rank_corrs.mean())
    print("Std daily rank correlation:", std_dev)

    return float(daily_rank_corrs.mean() / std_dev)


def score(solution: pd.DataFrame, submission: pd.DataFrame, row_id_column_name: str) -> float:
    solution = solution.copy()
    submission = submission.copy()

    del solution[row_id_column_name]
    del submission[row_id_column_name]

    assert all(solution.columns == submission.columns), "Solution and submission columns do not match."

    submission = submission.rename(
        columns={col: col.replace("target_", "prediction_") for col in submission.columns}
    )

    solution = solution.replace(SOLUTION_NULL_FILLER, None)

    merged = pd.concat([solution, submission], axis="columns")

    return rank_correlation_sharpe_ratio(merged)


def evaluate_submission(submission_df, solution_df, target_cols, name="model"):
    solution_eval = solution_df[["date_id"] + target_cols].copy()
    submission_eval = submission_df[["date_id"] + target_cols].copy()

    solution_eval = solution_eval.sort_values("date_id").reset_index(drop=True)
    submission_eval = submission_eval.sort_values("date_id").reset_index(drop=True)

    print("\n" + "=" * 70)
    print(f"Evaluating: {name}")
    print("=" * 70)
    print("Date IDs match:", list(solution_eval["date_id"]) == list(submission_eval["date_id"]))
    print("Columns match:", list(solution_eval.columns) == list(submission_eval.columns))

    model_score = score(
        solution=solution_eval,
        submission=submission_eval,
        row_id_column_name="date_id"
    )

    print(f"\n{name} score:", model_score)
    return model_score


# ============================================================
# 4. Helper functions for target-specific features
# ============================================================

def parse_pair_columns(pair_string):
    """
    Converts target_pairs pair strings into component column names.
    Examples:
    'US_Stock_VT_adj_close' -> ['US_Stock_VT_adj_close']
    'LME_PB_Close - US_Stock_VT_adj_close' -> ['LME_PB_Close', 'US_Stock_VT_adj_close']
    """
    if pd.isna(pair_string):
        return []

    return [p.strip() for p in str(pair_string).split(" - ")]


def safe_existing_columns(cols, train_df, test_df):
    """
    Keep only columns that exist in both train and test.
    """
    return [c for c in cols if c in train_df.columns and c in test_df.columns]


def build_target_specific_features(target):
    """
    Build safe target-specific features for one target.

    Features:
    - Pair asset raw values
    - Safe lags: 1, 2, 3, 5, 10
    - Safe rolling means/stds using shift(1)
    - Safe diffs: current - previous values
    - Pair spread features if target pair has two assets
    - Corrected lag-label feature using shift(lag + 1) for train
    """

    pair_string = target_pair_map[target]
    pair_cols = parse_pair_columns(pair_string)
    pair_cols = safe_existing_columns(pair_cols, train, test)

    # Fallback if pair cols are missing: use a small set of broad clean features
    if len(pair_cols) == 0:
        pair_cols = clean_feature_cols[:30]

    # Add a few broad market features so the model has context
    # Keep this small to avoid noisy huge feature sets.
    broad_context_cols = clean_feature_cols[:50]

    selected_cols = list(dict.fromkeys(pair_cols + broad_context_cols))

    combined = pd.concat(
        [
            train[["date_id"] + selected_cols],
            test[["date_id"] + selected_cols]
        ],
        axis=0,
        ignore_index=True
    ).sort_values("date_id").reset_index(drop=True)

    feature_frames = []

    # Raw selected columns
    raw_features = combined[selected_cols].copy()
    feature_frames.append(raw_features)

    # Safe lags and rolling features
    lags = [1, 2, 3, 5, 10]
    windows = [3, 5, 10, 20]
    diffs = [1, 2, 5]

    lag_dict = {}

    for col in selected_cols:
        for lag in lags:
            lag_dict[f"{col}_lag_{lag}"] = combined[col].shift(lag)

        shifted = combined[col].shift(1)

        for window in windows:
            lag_dict[f"{col}_roll_mean_{window}"] = shifted.rolling(window).mean()
            lag_dict[f"{col}_roll_std_{window}"] = shifted.rolling(window).std()

        for diff in diffs:
            lag_dict[f"{col}_diff_{diff}"] = combined[col].diff(diff)

    engineered = pd.DataFrame(lag_dict)
    feature_frames.append(engineered)

    # Pair spread features if two valid pair columns exist
    if len(pair_cols) == 2:
        a, b = pair_cols
        spread = combined[a] - combined[b]

        spread_dict = {
            f"{target}_spread": spread,
            f"{target}_spread_lag_1": spread.shift(1),
            f"{target}_spread_lag_2": spread.shift(2),
            f"{target}_spread_lag_3": spread.shift(3),
            f"{target}_spread_lag_5": spread.shift(5),
            f"{target}_spread_diff_1": spread.diff(1),
            f"{target}_spread_diff_2": spread.diff(2),
            f"{target}_spread_roll_mean_5": spread.shift(1).rolling(5).mean(),
            f"{target}_spread_roll_std_5": spread.shift(1).rolling(5).std(),
            f"{target}_spread_roll_mean_10": spread.shift(1).rolling(10).mean(),
            f"{target}_spread_roll_std_10": spread.shift(1).rolling(10).std(),
        }

        feature_frames.append(pd.DataFrame(spread_dict))

    # Combine features
    combined_features = pd.concat(feature_frames, axis=1)
    combined_features = combined_features.replace([np.inf, -np.inf], np.nan).copy()

    # Add date_id back
    combined_features.insert(0, "date_id", combined["date_id"].values)

    # Split train/test
    train_target_features = combined_features.iloc[:len(train)].copy()
    test_target_features = combined_features.iloc[len(train):].copy()

    # Add corrected lag-label feature for this target
    target_lag = int(target_lag_map[target])

    train_target_features[f"lag_label_{target}"] = train_labels[target].shift(target_lag + 1).values

    # For test, pull from correct lag file
    lag_file = f"{competition_data_path}/lagged_test_labels/test_labels_lag_{target_lag}.csv"
    lag_df = pd.read_csv(lag_file)

    if target in lag_df.columns:
        test_lag_feature = lag_df[["date_id", target]].copy()
        test_lag_feature = test_lag_feature.rename(columns={target: f"lag_label_{target}"})

        test_target_features = test_target_features.merge(
            test_lag_feature,
            on="date_id",
            how="left"
        )
    else:
        test_target_features[f"lag_label_{target}"] = np.nan

    # Final feature columns
    target_feature_cols = [c for c in train_target_features.columns if c != "date_id"]

    X_train_target_all = train_target_features[target_feature_cols]
    X_test_target = test_target_features[target_feature_cols]

    return X_train_target_all, X_test_target


# ============================================================
# 5. Train target-specific XGBoost models
# ============================================================

train_nonleaky_mask = train["date_id"] < local_test_start

pair_model_params = {
    "n_estimators": 250,
    "learning_rate": 0.04,
    "max_depth": 2,
    "subsample": 0.9,
    "colsample_bytree": 0.85,
    "reg_alpha": 0.05,
    "reg_lambda": 4.0,
    "min_child_weight": 4,
}

submission_pair_xgb = pd.DataFrame(index=test.index)

start_time = time.time()

for i, target in enumerate(target_cols):
    print(f"\nTraining target-specific model {i + 1}/{len(target_cols)}: {target}")

    X_train_target_all, X_test_target = build_target_specific_features(target)

    X_train_target = X_train_target_all.loc[train_nonleaky_mask]

    y_train = train_labels.loc[train_nonleaky_mask, target].values
    target_mask = ~np.isnan(y_train)

    if target_mask.sum() < 100:
        print(f"Skipping {target}: not enough non-missing labels.")
        submission_pair_xgb[target] = 0.0
        continue

    model = XGBRegressor(
        **pair_model_params,
        objective="reg:squarederror",
        tree_method="hist",
        random_state=42,
        n_jobs=-1,
        eval_metric="rmse"
    )

    model.fit(
        X_train_target.loc[target_mask],
        y_train[target_mask],
        verbose=False
    )

    submission_pair_xgb[target] = model.predict(X_test_target)

    if (i + 1) % 25 == 0:
        elapsed = (time.time() - start_time) / 60
        print("\n" + "-" * 60)
        print(f"Finished {i + 1}/{len(target_cols)} targets in {elapsed:.2f} minutes")
        print("-" * 60)

    del model, X_train_target_all, X_train_target, X_test_target
    gc.collect()

submission_pair_xgb.insert(0, "date_id", test["date_id"].values)
submission_pair_xgb = submission_pair_xgb[["date_id"] + target_cols].copy()

print("\nTarget-specific XGBoost prediction shape:", submission_pair_xgb.shape)
display(submission_pair_xgb.head())

submission_pair_xgb.to_parquet("/kaggle/working/pair_specific_xgb.parquet", index=False)


# ============================================================
# 6. Evaluate target-specific model
# ============================================================

pair_xgb_score = evaluate_submission(
    submission_pair_xgb,
    solution,
    target_cols,
    name="Target-specific safe XGBoost"
)

print("\nTarget-specific XGBoost score:")
print(pair_xgb_score)


# ============================================================
# 7. Optional blend with existing XGBoost result if available
# ============================================================
# This assumes you have a previous strong XGBoost prediction DataFrame called submission_xgb.
# If you do not have it in memory, this section will skip.

if "submission_xgb" in globals():
    print("\nFound existing submission_xgb. Running rank blend...")

    xgb_eval = submission_xgb.sort_values("date_id").reset_index(drop=True)
    pair_eval = submission_pair_xgb.sort_values("date_id").reset_index(drop=True)

    xgb_rank = xgb_eval[target_cols].rank(axis=1, method="average", ascending=True)
    pair_rank = pair_eval[target_cols].rank(axis=1, method="average", ascending=True)

    blend_scores = {}
    blend_submissions = {}

    for w_xgb in [0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80, 0.90]:
        w_pair = 1.0 - w_xgb

        blended_values = (w_xgb * xgb_rank) + (w_pair * pair_rank)

        blended_submission = blended_values.copy()
        blended_submission.insert(0, "date_id", xgb_eval["date_id"].values)
        blended_submission = blended_submission[["date_id"] + target_cols].copy()

        blend_name = f"oldxgb_{w_xgb:.2f}_pairxgb_{w_pair:.2f}"

        blend_score = evaluate_submission(
            blended_submission,
            solution,
            target_cols,
            name=blend_name
        )

        blend_scores[blend_name] = blend_score
        blend_submissions[blend_name] = blended_submission

    blend_scores_series = pd.Series(blend_scores).sort_values(ascending=False)

    print("\n" + "=" * 70)
    print("Blend results: old XGBoost + target-specific XGBoost")
    print("=" * 70)
    display(blend_scores_series)

    best_blend_name = blend_scores_series.index[0]
    best_blend_submission = blend_submissions[best_blend_name]

    print("Best blend:", best_blend_name)
    print("Best blend score:", blend_scores_series.iloc[0])

    best_blend_submission.to_parquet("/kaggle/working/submission_pair_blend.parquet", index=False)
    best_blend_submission.to_parquet("/kaggle/working/submission.parquet", index=False)

else:
    print("\nNo existing submission_xgb found in memory.")
    print("Saving pair-specific XGBoost as submission.parquet.")
    submission_pair_xgb.to_parquet("/kaggle/working/submission.parquet", index=False)


print("\nFiles in /kaggle/working:")
print(os.listdir("/kaggle/working"))
```

```python
# ============================================================
# LightGBM Setup: Non-Leaky + Corrected Lag Label Features
# Trains 424 target models and blends with existing XGBoost
# ============================================================

import os
import gc
import time
import warnings
import numpy as np
import pandas as pd

warnings.filterwarnings("ignore", category=pd.errors.PerformanceWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

from lightgbm import LGBMRegressor

SOLUTION_NULL_FILLER = -999999

competition_data_path = "/kaggle/input/competitions/mitsui-commodity-prediction-challenge"
ground_truth_path = "/kaggle/input/datasets/stephenweiler208/ground-truth"

# ============================================================
# 1. Load ground truth
# ============================================================

if os.path.isdir(ground_truth_path):
    csv_files = [f for f in os.listdir(ground_truth_path) if f.endswith(".csv")]
    if len(csv_files) == 0:
        raise FileNotFoundError("No CSV found in ground truth folder.")
    ground_truth_path = os.path.join(ground_truth_path, csv_files[0])

solution = pd.read_csv(ground_truth_path)

local_test_start = solution["date_id"].min()
local_test_end = solution["date_id"].max()

print("Ground truth shape:", solution.shape)
print("Ground truth date range:", local_test_start, "to", local_test_end)

# ============================================================
# 2. Target metadata
# ============================================================

target_pairs = pd.read_csv(f"{competition_data_path}/target_pairs.csv")
target_cols = target_pairs["target"].tolist()
target_lag_map = target_pairs.set_index("target")["lag"].to_dict()

print("Number of targets:", len(target_cols))
display(target_pairs.head())

# ============================================================
# 3. Clean base features
# ============================================================

id_cols = ["date_id"]
feature_cols = [c for c in train.columns if c not in id_cols]

feature_missing_rate = train[feature_cols].isna().mean()
clean_feature_cols = feature_missing_rate[feature_missing_rate <= 0.50].index.tolist()

print("Original feature count:", len(feature_cols))
print("Clean feature count:", len(clean_feature_cols))

# ============================================================
# 4. Build corrected lag-label features for TRAIN
# No fragmentation: dictionary -> DataFrame
# ============================================================

lag_feature_dict = {}

for target in target_cols:
    lag = int(target_lag_map[target])
    lag_feature_dict[f"lag_label_{target}"] = train_labels[target].shift(lag + 1)

train_label_lag_features = pd.DataFrame(lag_feature_dict)
train_label_lag_features.insert(0, "date_id", train_labels["date_id"].values)
train_label_lag_features = train_label_lag_features.copy()

print("Train lag-label feature shape:", train_label_lag_features.shape)
display(train_label_lag_features.head())

# ============================================================
# 5. Build lag-label features for TEST
# ============================================================

test_label_lag_features = pd.DataFrame({"date_id": test["date_id"].values})

for lag in [1, 2, 3, 4]:
    lag_df = pd.read_csv(
        f"{competition_data_path}/lagged_test_labels/test_labels_lag_{lag}.csv"
    )

    lag_targets = target_pairs.loc[target_pairs["lag"] == lag, "target"].tolist()
    available_targets = [c for c in lag_targets if c in lag_df.columns]

    temp = lag_df[["date_id"] + available_targets].copy()
    temp = temp.rename(columns={c: f"lag_label_{c}" for c in available_targets})

    test_label_lag_features = test_label_lag_features.merge(
        temp,
        on="date_id",
        how="left"
    )

test_label_lag_features = test_label_lag_features.copy()

print("Test lag-label feature shape:", test_label_lag_features.shape)
display(test_label_lag_features.head())

# ============================================================
# 6. Build final feature matrices
# ============================================================

train_features_lgbm = train[["date_id"] + clean_feature_cols].merge(
    train_label_lag_features,
    on="date_id",
    how="left"
)

test_features_lgbm = test[["date_id"] + clean_feature_cols].merge(
    test_label_lag_features,
    on="date_id",
    how="left"
)

train_features_lgbm = train_features_lgbm.replace([np.inf, -np.inf], np.nan).copy()
test_features_lgbm = test_features_lgbm.replace([np.inf, -np.inf], np.nan).copy()

lgbm_feature_cols = [c for c in train_features_lgbm.columns if c != "date_id"]

train_nonleaky_mask = train_features_lgbm["date_id"] < local_test_start

X_train_lgbm = train_features_lgbm.loc[train_nonleaky_mask, lgbm_feature_cols].copy()
X_test_lgbm = test_features_lgbm[lgbm_feature_cols].copy()

print("LightGBM feature count:", len(lgbm_feature_cols))
print("Non-leaky training rows:", X_train_lgbm.shape)
print("Local test rows:", X_test_lgbm.shape)
print(
    "Training date range:",
    train_features_lgbm.loc[train_nonleaky_mask, "date_id"].min(),
    "to",
    train_features_lgbm.loc[train_nonleaky_mask, "date_id"].max()
)

# ============================================================
# 7. Metric functions
# ============================================================

def rank_correlation_sharpe_ratio(merged_df: pd.DataFrame) -> float:
    prediction_cols = [col for col in merged_df.columns if col.startswith("prediction_")]
    target_cols_metric = [col for col in merged_df.columns if col.startswith("target_")]

    daily_rank_corrs = []

    for _, row in merged_df.iterrows():
        non_null_targets = [col for col in target_cols_metric if not pd.isnull(row[col])]

        matching_predictions = [
            col for col in prediction_cols
            if col.replace("prediction", "target") in non_null_targets
        ]

        if len(non_null_targets) < 3:
            daily_rank_corrs.append(np.nan)
            continue

        if row[non_null_targets].std(ddof=0) == 0 or row[matching_predictions].std(ddof=0) == 0:
            daily_rank_corrs.append(np.nan)
            continue

        corr = np.corrcoef(
            row[matching_predictions].rank(method="average"),
            row[non_null_targets].rank(method="average")
        )[0, 1]

        daily_rank_corrs.append(corr)

    daily_rank_corrs = pd.Series(daily_rank_corrs).dropna()

    std_dev = daily_rank_corrs.std(ddof=0)

    if std_dev == 0:
        raise ZeroDivisionError("Denominator is zero, unable to compute Sharpe ratio.")

    print("\nDaily rank correlation summary:")
    display(daily_rank_corrs.describe())

    print("\nMean daily rank correlation:", daily_rank_corrs.mean())
    print("Std daily rank correlation:", std_dev)

    return float(daily_rank_corrs.mean() / std_dev)


def score(solution: pd.DataFrame, submission: pd.DataFrame, row_id_column_name: str) -> float:
    solution = solution.copy()
    submission = submission.copy()

    del solution[row_id_column_name]
    del submission[row_id_column_name]

    assert all(solution.columns == submission.columns), "Solution and submission columns do not match."

    submission = submission.rename(
        columns={col: col.replace("target_", "prediction_") for col in submission.columns}
    )

    solution = solution.replace(SOLUTION_NULL_FILLER, None)

    merged = pd.concat([solution, submission], axis="columns")

    return rank_correlation_sharpe_ratio(merged)


def evaluate_submission(submission_df, solution_df, target_cols, name="model"):
    solution_eval = solution_df[["date_id"] + target_cols].copy()
    submission_eval = submission_df[["date_id"] + target_cols].copy()

    solution_eval = solution_eval.sort_values("date_id").reset_index(drop=True)
    submission_eval = submission_eval.sort_values("date_id").reset_index(drop=True)

    print("\n" + "=" * 70)
    print(f"Evaluating: {name}")
    print("=" * 70)
    print("Date IDs match:", list(solution_eval["date_id"]) == list(submission_eval["date_id"]))
    print("Columns match:", list(solution_eval.columns) == list(submission_eval.columns))

    model_score = score(
        solution=solution_eval,
        submission=submission_eval,
        row_id_column_name="date_id"
    )

    print(f"\n{name} score:", model_score)

    return model_score

# ============================================================
# 8. Train LightGBM models
# ============================================================

lgbm_params = {
    "n_estimators": 400,
    "learning_rate": 0.025,
    "max_depth": 2,
    "num_leaves": 7,
    "min_child_samples": 80,
    "subsample": 0.9,
    "subsample_freq": 1,
    "colsample_bytree": 0.8,
    "reg_alpha": 0.1,
    "reg_lambda": 5.0,
    "random_state": 42,
    "n_jobs": -1,
    "verbose": -1,
}

submission_lgbm = pd.DataFrame(index=test.index)

start_time = time.time()

for i, target in enumerate(target_cols):
    y_train = train_labels.loc[train_nonleaky_mask, target].values
    target_mask = ~np.isnan(y_train)

    if target_mask.sum() < 100:
        submission_lgbm[target] = 0.0
        continue

    model = LGBMRegressor(**lgbm_params)

    model.fit(
        X_train_lgbm.loc[target_mask],
        y_train[target_mask]
    )

    submission_lgbm[target] = model.predict(X_test_lgbm)

    if (i + 1) % 25 == 0:
        elapsed = (time.time() - start_time) / 60
        print(f"Finished {i + 1}/{len(target_cols)} LightGBM targets in {elapsed:.2f} minutes")

    del model
    gc.collect()

submission_lgbm.insert(0, "date_id", test["date_id"].values)
submission_lgbm = submission_lgbm[["date_id"] + target_cols].copy()

print("\nLightGBM prediction shape:", submission_lgbm.shape)
display(submission_lgbm.head())

submission_lgbm.to_parquet("/kaggle/working/lgbm_corrected_laglabel.parquet", index=False)

# ============================================================
# 9. Evaluate LightGBM alone
# ============================================================

lgbm_score = evaluate_submission(
    submission_lgbm,
    solution,
    target_cols,
    name="LightGBM corrected lag-label"
)

print("\nLightGBM score:", lgbm_score)

# ============================================================
# 10. Blend with existing XGBoost if available
# ============================================================

# Try to find XGBoost predictions from memory or saved file
if "submission_xgb" in globals():
    print("\nFound submission_xgb in memory.")
    xgb_pred = submission_xgb.copy()

elif os.path.exists("/kaggle/working/xgb_corrected_laglabel.parquet"):
    print("\nLoading XGBoost predictions from /kaggle/working/xgb_corrected_laglabel.parquet")
    xgb_pred = pd.read_parquet("/kaggle/working/xgb_corrected_laglabel.parquet")

elif os.path.exists("/kaggle/working/submission_best_local_blend.parquet"):
    print("\nLoading previous best XGBoost/blend predictions from /kaggle/working/submission_best_local_blend.parquet")
    xgb_pred = pd.read_parquet("/kaggle/working/submission_best_local_blend.parquet")

else:
    xgb_pred = None
    print("\nNo existing XGBoost prediction file found. Skipping blend.")

if xgb_pred is not None:
    xgb_pred = xgb_pred[["date_id"] + target_cols].copy()

    xgb_score = evaluate_submission(
        xgb_pred,
        solution,
        target_cols,
        name="Existing XGBoost / previous best"
    )

    xgb_eval = xgb_pred.sort_values("date_id").reset_index(drop=True)
    lgbm_eval = submission_lgbm.sort_values("date_id").reset_index(drop=True)

    xgb_rank = xgb_eval[target_cols].rank(axis=1, method="average", ascending=True)
    lgbm_rank = lgbm_eval[target_cols].rank(axis=1, method="average", ascending=True)

    blend_scores = {}
    blend_submissions = {}

    for w_xgb in [0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80, 0.90]:
        w_lgbm = 1.0 - w_xgb

        blended_values = (w_xgb * xgb_rank) + (w_lgbm * lgbm_rank)

        blended_submission = blended_values.copy()
        blended_submission.insert(0, "date_id", xgb_eval["date_id"].values)
        blended_submission = blended_submission[["date_id"] + target_cols].copy()

        blend_name = f"xgb_{w_xgb:.2f}_lgbm_{w_lgbm:.2f}"

        blend_score = evaluate_submission(
            blended_submission,
            solution,
            target_cols,
            name=blend_name
        )

        blend_scores[blend_name] = blend_score
        blend_submissions[blend_name] = blended_submission

    blend_scores_series = pd.Series(blend_scores).sort_values(ascending=False)

    print("\n" + "=" * 70)
    print("Blend results: XGBoost + LightGBM")
    print("=" * 70)
    display(blend_scores_series)

    best_blend_name = blend_scores_series.index[0]
    best_blend_score = blend_scores_series.iloc[0]
    best_blend_submission = blend_submissions[best_blend_name]

    print("\nBest blend:", best_blend_name)
    print("Best blend score:", best_blend_score)

    best_blend_submission.to_parquet("/kaggle/working/submission_xgb_lgbm_blend.parquet", index=False)
    best_blend_submission.to_parquet("/kaggle/working/submission.parquet", index=False)

else:
    submission_lgbm.to_parquet("/kaggle/working/submission.parquet", index=False)

print("\nFiles in /kaggle/working:")
print(os.listdir("/kaggle/working"))
```

```python
# ============================================================
# Self-contained ensemble:
# Your Random Forest output + Friend's LightGBM output
# ============================================================

import os
import numpy as np
import pandas as pd

# ============================================================
# 1. Paths
# ============================================================

COMPETITION_PATH = "/kaggle/input/competitions/mitsui-commodity-prediction-challenge"
GROUND_TRUTH_PATH = "/kaggle/input/datasets/stephenweiler208/ground-truth"

# Friend's LightGBM output folder
lgbm_path = "/kaggle/input/datasets/stephenweiler208/lgbm-adam"

# Your Random Forest output
# Change this if your RF file has a different name
possible_rf_paths = "/kaggle/input/datasets/stephenweiler208/random-forest1"

# ============================================================
# 2. Helper: load a parquet/csv from file or folder
# ============================================================

def load_prediction_file(path):
    """
    Loads either:
    - a .parquet file
    - a .csv file
    - a folder containing one .parquet or .csv file
    """

    if not os.path.exists(path):
        raise FileNotFoundError(f"Path does not exist: {path}")

    if os.path.isdir(path):
        files = os.listdir(path)
        print(f"\nFiles in folder {path}:")
        print(files)

        parquet_files = [f for f in files if f.endswith(".parquet")]
        csv_files = [f for f in files if f.endswith(".csv")]

        if len(parquet_files) > 0:
            file_path = os.path.join(path, parquet_files[0])
            print("Loading parquet:", file_path)
            return pd.read_parquet(file_path)

        if len(csv_files) > 0:
            file_path = os.path.join(path, csv_files[0])
            print("Loading csv:", file_path)
            return pd.read_csv(file_path)

        raise FileNotFoundError(f"No .parquet or .csv file found in folder: {path}")

    if path.endswith(".parquet"):
        print("Loading parquet:", path)
        return pd.read_parquet(path)

    if path.endswith(".csv"):
        print("Loading csv:", path)
        return pd.read_csv(path)

    raise ValueError(f"Unsupported file type or folder: {path}")


def find_existing_path(paths):
    for path in paths:
        if os.path.exists(path):
            return path
    raise FileNotFoundError(
        "None of the possible RF paths exist. Check /kaggle/working or upload the RF file as a dataset."
    )


# ============================================================
# 3. Load competition metadata and ground truth
# ============================================================

target_pairs = pd.read_csv(f"{COMPETITION_PATH}/target_pairs.csv")
target_cols = target_pairs["target"].tolist()

test = pd.read_csv(f"{COMPETITION_PATH}/test.csv")
test_date_ids = test["date_id"].values

if os.path.isdir(GROUND_TRUTH_PATH):
    gt_files = os.listdir(GROUND_TRUTH_PATH)
    print("\nFiles in ground truth folder:")
    print(gt_files)

    gt_csvs = [f for f in gt_files if f.endswith(".csv")]

    if len(gt_csvs) == 0:
        raise FileNotFoundError("No CSV found in ground truth folder.")

    ground_truth_file = os.path.join(GROUND_TRUTH_PATH, gt_csvs[0])
else:
    ground_truth_file = GROUND_TRUTH_PATH

solution = pd.read_csv(ground_truth_file)

print("\nLoaded metadata:")
print("target_pairs:", target_pairs.shape)
print("test:", test.shape)
print("solution:", solution.shape)
print("Number of targets:", len(target_cols))

# ============================================================
# 4. Load RF and friend's LGBM predictions
# ============================================================

rf_path = possible_rf_paths

rf_pred = load_prediction_file(rf_path)
friend_lgbm_pred = load_prediction_file(friend_lgbm_path)

print("\nRaw prediction shapes:")
print("RF:", rf_pred.shape)
print("Friend LGBM:", friend_lgbm_pred.shape)

display(rf_pred.head())
display(friend_lgbm_pred.head())

# ============================================================
# 5. Add date_id if missing
# ============================================================

if "date_id" not in rf_pred.columns:
    rf_pred = rf_pred.copy()
    rf_pred.insert(0, "date_id", test_date_ids[:len(rf_pred)])

if "date_id" not in friend_lgbm_pred.columns:
    friend_lgbm_pred = friend_lgbm_pred.copy()
    friend_lgbm_pred.insert(0, "date_id", test_date_ids[:len(friend_lgbm_pred)])

# ============================================================
# 6. Align rows and columns
# ============================================================

# Only keep target columns that appear in all three files
available_targets = [
    c for c in target_cols
    if c in rf_pred.columns and c in friend_lgbm_pred.columns and c in solution.columns
]

if len(available_targets) != len(target_cols):
    print("Warning: not all target columns were found.")
    print("Expected:", len(target_cols))
    print("Available:", len(available_targets))

target_cols = available_targets

rf_pred = rf_pred[["date_id"] + target_cols].sort_values("date_id").reset_index(drop=True)
friend_lgbm_pred = friend_lgbm_pred[["date_id"] + target_cols].sort_values("date_id").reset_index(drop=True)
solution = solution[["date_id"] + target_cols].sort_values("date_id").reset_index(drop=True)

# Keep only dates common to all three
common_dates = sorted(
    set(rf_pred["date_id"])
    & set(friend_lgbm_pred["date_id"])
    & set(solution["date_id"])
)

rf_pred = rf_pred[rf_pred["date_id"].isin(common_dates)].sort_values("date_id").reset_index(drop=True)
friend_lgbm_pred = friend_lgbm_pred[friend_lgbm_pred["date_id"].isin(common_dates)].sort_values("date_id").reset_index(drop=True)
solution = solution[solution["date_id"].isin(common_dates)].sort_values("date_id").reset_index(drop=True)

print("\nAligned shapes:")
print("RF:", rf_pred.shape)
print("Friend LGBM:", friend_lgbm_pred.shape)
print("Solution:", solution.shape)

print("\nDate ranges:")
print("RF:", rf_pred["date_id"].min(), "to", rf_pred["date_id"].max())
print("Friend LGBM:", friend_lgbm_pred["date_id"].min(), "to", friend_lgbm_pred["date_id"].max())
print("Solution:", solution["date_id"].min(), "to", solution["date_id"].max())

assert list(rf_pred["date_id"]) == list(friend_lgbm_pred["date_id"]), "RF and LGBM date_id mismatch"
assert list(rf_pred["date_id"]) == list(solution["date_id"]), "Prediction and solution date_id mismatch"
assert list(rf_pred.columns) == list(friend_lgbm_pred.columns), "Prediction columns mismatch"
assert list(rf_pred.columns) == list(solution.columns), "Solution columns mismatch"

# ============================================================
# 7. Competition-style scoring functions
# ============================================================

SOLUTION_NULL_FILLER = -999999

def rank_correlation_sharpe_ratio(merged_df: pd.DataFrame) -> float:
    prediction_cols = [col for col in merged_df.columns if col.startswith("prediction_")]
    target_cols_metric = [col for col in merged_df.columns if col.startswith("target_")]

    daily_rank_corrs = []

    for _, row in merged_df.iterrows():
        non_null_targets = [col for col in target_cols_metric if not pd.isnull(row[col])]

        matching_predictions = [
            col for col in prediction_cols
            if col.replace("prediction", "target") in non_null_targets
        ]

        if len(non_null_targets) < 3:
            daily_rank_corrs.append(np.nan)
            continue

        if row[non_null_targets].std(ddof=0) == 0 or row[matching_predictions].std(ddof=0) == 0:
            daily_rank_corrs.append(np.nan)
            continue

        corr = np.corrcoef(
            row[matching_predictions].rank(method="average"),
            row[non_null_targets].rank(method="average")
        )[0, 1]

        daily_rank_corrs.append(corr)

    daily_rank_corrs = pd.Series(daily_rank_corrs).dropna()

    std_dev = daily_rank_corrs.std(ddof=0)

    if std_dev == 0:
        raise ZeroDivisionError("Daily rank correlation standard deviation is zero.")

    print("\nDaily rank correlation summary:")
    display(daily_rank_corrs.describe())

    print("Mean daily rank correlation:", daily_rank_corrs.mean())
    print("Std daily rank correlation:", std_dev)

    return float(daily_rank_corrs.mean() / std_dev)


def score(solution_df: pd.DataFrame, submission_df: pd.DataFrame, row_id_column_name: str) -> float:
    solution_df = solution_df.copy()
    submission_df = submission_df.copy()

    del solution_df[row_id_column_name]
    del submission_df[row_id_column_name]

    assert all(solution_df.columns == submission_df.columns), "Solution and submission columns do not match."

    submission_df = submission_df.rename(
        columns={col: col.replace("target_", "prediction_") for col in submission_df.columns}
    )

    solution_df = solution_df.replace(SOLUTION_NULL_FILLER, None)

    merged_df = pd.concat([solution_df, submission_df], axis="columns")

    return rank_correlation_sharpe_ratio(merged_df)


def evaluate_submission(submission_df, solution_df, target_cols, name="model"):
    solution_eval = solution_df[["date_id"] + target_cols].copy()
    submission_eval = submission_df[["date_id"] + target_cols].copy()

    solution_eval = solution_eval.sort_values("date_id").reset_index(drop=True)
    submission_eval = submission_eval.sort_values("date_id").reset_index(drop=True)

    assert list(solution_eval["date_id"]) == list(submission_eval["date_id"]), "date_id mismatch"
    assert list(solution_eval.columns) == list(submission_eval.columns), "column mismatch"

    model_score = score(
        solution_df=solution_eval,
        submission_df=submission_eval,
        row_id_column_name="date_id"
    )

    print(f"\n{name} score:", model_score)

    return model_score

# ============================================================
# 8. Score individual models
# ============================================================

rf_score = evaluate_submission(
    rf_pred,
    solution,
    target_cols,
    name="Your Random Forest"
)

friend_lgbm_score = evaluate_submission(
    friend_lgbm_pred,
    solution,
    target_cols,
    name="Friend LightGBM"
)

print("\nIndividual scores:")
print("RF:", rf_score)
print("Friend LGBM:", friend_lgbm_score)

# ============================================================
# 9. Rank-blend RF + friend's LGBM
# ============================================================

rf_rank = rf_pred[target_cols].rank(axis=1, method="average", ascending=True)
friend_lgbm_rank = friend_lgbm_pred[target_cols].rank(axis=1, method="average", ascending=True)

blend_scores = {}
blend_submissions = {}

for w_lgbm in [0.05, 0.10, 0.20, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80, 0.90, 0.95]:
    w_rf = 1.0 - w_lgbm

    blended_values = (w_lgbm * friend_lgbm_rank) + (w_rf * rf_rank)

    blended_submission = blended_values.copy()
    blended_submission.insert(0, "date_id", rf_pred["date_id"].values)
    blended_submission = blended_submission[["date_id"] + target_cols].copy()

    blend_name = f"friendlgbm_{w_lgbm:.2f}_rf_{w_rf:.2f}"

    blend_score = evaluate_submission(
        blended_submission,
        solution,
        target_cols,
        name=blend_name
    )

    blend_scores[blend_name] = blend_score
    blend_submissions[blend_name] = blended_submission

blend_scores_series = pd.Series(blend_scores).sort_values(ascending=False)

print("\nRF + Friend LGBM blend results:")
display(blend_scores_series)

best_blend_name = blend_scores_series.index[0]
best_blend_score = blend_scores_series.iloc[0]
best_blend_submission = blend_submissions[best_blend_name]

print("\nBest blend:", best_blend_name)
print("Best blend score:", best_blend_score)

# ============================================================
# 10. Save best blend
# ============================================================

best_blend_submission.to_parquet("/kaggle/working/submission_rf_friend_lgbm_blend.parquet", index=False)
best_blend_submission.to_parquet("/kaggle/working/submission.parquet", index=False)

print("\nSaved:")
print("/kaggle/working/submission_rf_friend_lgbm_blend.parquet")
print("/kaggle/working/submission.parquet")
```
